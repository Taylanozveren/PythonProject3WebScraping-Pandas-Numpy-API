You have been hired as a junior Data Engineer and are tasked with creating a script that can extract the list of the top 10 largest economies of the world in descending order of their GDPs in Billion USD (rounded to 2 decimal places), as logged by the International Monetary Fund (IMF).

### Exercise 1
#### Extract the required GDP data from the given URL using Web Scraping.
##### We need to extract countries' IMF-linked GDP data, and only the top 10 countries are required
##### Edit column names

```python
URL = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'

# Pandas kütüphanesini kullanarak web sayfasından tabloları çıkarın ve 3. tabloyu alın.
tables = pd.read_html(URL)
df = tables[3] # ülkelerin GDP&IMF verilerini içeren tablo

# Sütun başlıklarını sütun numaraları ile değiştirin
df.columns = range(df.shape[1]) # DataFrame'in sütun sayısını döndürür ve bu değer sütun başlıklarını yeniden adlandırmak için kullanılır.

# Sadece ülke ismi ve GDP değerlerini içeren sütunları alın 
df = df[[0,2]]

# İlk 10 ekonomiyi temsil eden satırları seçin
df = df.iloc[1:11,:]

# Sütun adlarını "Country" ve "GDP (Million USD)" olarak atayın
df.columns = ['Country', 'GDP (Million USD)']
df
```

### Exercise 2
#### Modify the GDP column of the DataFrame, converting the value available in Million USD to Billion USD. Use the round() method of Numpy library to round the value to 2 decimal places. Modify the header of the DataFrame to GDP (Billion USD).

```python

# Örnek DataFrame
df = pd.DataFrame({
    'Country': ['Country A', 'Country B', 'Country C'],
    'GDP (Million USD)': [1500000, 1200000, 900000]
})

# 1. 'GDP (Million USD)' sütununun veri tipini integer olarak değiştirin
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)

# 2. GDP değerini Milyon USD'den Milyar USD'ye dönüştürün
df[['GDP (Million USD)']] = df[['GDP (Million USD)']] / 1000

# 3. numpy.round() yöntemini kullanarak değeri 2 ondalık basamağa yuvarlayın
df[['GDP (Million USD)']] = np.round(df[['GDP (Million USD)']], 2)

# 4. Sütun başlığını 'GDP (Million USD)''den 'GDP (Billion USD)''ye değiştirin
df.rename(columns={'GDP (Million USD)': 'GDP (Billion USD)'}, inplace=True)

# Sonuçları kontrol edin
print(df)
```
